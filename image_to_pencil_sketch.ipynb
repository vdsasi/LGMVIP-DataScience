```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

image = plt.imread('flower.jpg')

fig, ax = plt.subplots()
ax.set_title('Original Image')
ax.imshow(image)
fig.set_size_inches(4, 4) 

plt.show()
```

```python
grey_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
inverted=255-grey_image
```

```python
bluerred=cv2.GaussianBlur(inverted,(21,21),0)
invertedblur=255-bluerred
plt.imshow(invertedblur)
```

```python
pencilsketch=cv2.divide(grey_image,invertedblur,scale=256.0)
```

```python
plt.imshow(pencilsketch, cmap = "gray")
```
